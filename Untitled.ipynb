```python
eventtime_rets = (
    final_data2.merge(fomc_rates, on = ["Date"], how = 'left', 
                      validate = 'm:1', indicator = True)
    
    .assign(event = lambda x: (x['_merge'] == 'both').astype(int),
                   increment = np.arange(len(final_data2)))
    
    .assign(inc_at_e = lambda x: x['event']*x['increment'])
    
    #inc_at_e alwasy equal to increment # at firms event
    .assign(inc_at_e = lambda x: x.groupby('Increase')['inc_at_e'].transform(sum))
    
    #compute event time
    .assign(event_time = lambda x: x['increment'] -x['inc_at_e'])
   
    #reduce event time [-15, +15]
    .query('(event_time >= -15) & (event_time <= 15)')
)

```

```python
rate_change_rets = (eventtime_rets
                    .assign(R_0_2 = lambda x: np.where(((x.event_time >= 0) & (x.event_time <=2)),
                                                        1+x['ret'],
                                                        np.nan),
                            R_2_10 = lambda x: np.where(((x.event_time >= 3) & (x.event_time <=10)),
                                                        1+x['ret'],
                                                        np.nan),
                           )
                    .groupby(['Increase'])
                    [['R_0_2','R_0_2']].prod()
                    -1
                   ).reset_index()
```
