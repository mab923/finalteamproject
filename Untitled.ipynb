```python
final_data = sp500_ret.merge(Vbilx, 
              on=['Date'],
              how='left',
              validate='one_to_one') # or 'many_to_one'
```

```python
final_data2 = final_data.merge(Vbirx, on=['Date'],
              how='left',
              validate='one_to_one')
```

### Create Dummy Variable

```python
fomc_rates 
fomc_rates['Change'] = np.where(fomc_rates['Increase'] > 0,1,0)
fomc_rates
```

**Dummy variable equals 1 when the rate increases, 0 if it was decreased.

### Create Event Time Variable

```python
event_time_rets = (
    final_data2.merge(fomc_rates, on = ["Date"], how = 'left', 
                      validate = 'm:1', indicator = True)
    
    .assign(event = lambda x: (x['_merge'] == 'both').astype(int),
                   increment = np.arange(len(final_data2)))
    
    .assign(inc_at_e = lambda x: x['event']*x['increment'])
    
    #inc_at_e alwasy equal to increment # at firms event
    .assign(inc_at_e = lambda x: x.groupby('Change')['inc_at_e'].transform(sum))
    
    #compute event time
    .assign(event_time = lambda x: x['increment'] -x['inc_at_e'])
   
    #reduce event time [-15, +15]
    .query('(event_time >= -15) & (event_time <= 15)')
)
```

```python
event_time_rets
```

```python
rate_change_rets = (eventtime_rets
                    .assign(R_0_2 = lambda x: np.where(((x.event_time >= 0) & (x.event_time <=2)),
                                                        1+x['sp500_ret'],
                                                        np.nan),
                            R_2_10 = lambda x: np.where(((x.event_time >= 3) & (x.event_time <=10)),
                                                        1+x['sp500_ret'],
                                                        np.nan),
                           )
                    .groupby(['sp500_ret', 'VBLIX_daily_ret', 'VBIRX_daily_ret', 'Increase', 'Decrease'])
                    [['R_0_2','R_0_2']].prod()
                    -1
                   ).reset_index()
```
